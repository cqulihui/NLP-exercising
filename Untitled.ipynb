## 制作词云图

```python
import jieba
from wordcloud import WordCloud,STOPWORDS,ImageColorGenerator
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import pymysql
from pandas import Series,DataFrame
import pandas as pd
```

```python
#连接数据库
db = pymysql.connect('10.51.218.74','lihui','lihui','joywok_lihui')
cursor = db.cursor()
```

```python
#读取数据库中的信息流
sqlpost = "SELECT DISTINCT content FROM 21_dogesoft_as"
cursor.execute(sqlpost)
resultpost = cursor.fetchall()
```

```python
dfpost = pd.DataFrame(list(resultpost),columns = ['content'])
```

```python
#将信息流中的姓名进行替换
sqlemployee = "SELECT * FROM 21_dogesoft_employee"
cursor.execute(sqlemployee)
resultemp = cursor.fetchall()
dfemployee = pd.DataFrame(list(resultemp),columns = ['uid','uname','pinyin','namesys'])
i = 0
while i < dfemployee.shape[0]:
    if dfemployee.iloc[i,3] is None:
        i = i + 1
    else:
        dfpost['content'] = dfpost['content'].apply(lambda x: x.replace(dfemployee.iloc[i,3],' '))
        i = i + 1
i = 0
while i < dfemployee.shape[0]:
    if dfemployee.iloc[i,1] is None:
        i = i + 1
    else:
        dfpost['content'] = dfpost['content'].apply(lambda x: x.replace(dfemployee.iloc[i,1],' '))
        i = i + 1
```

```python
#对信息流分词
i = 0
poststr = ''
while i < dfpost.shape[0]:
    poststr += ' '.join(jieba.cut(dfpost.iloc[i,0]))
    i = i + 1
```

```python
# 绘制信息流词云图
wc = WordCloud(background_color='white',max_words=200,max_font_size=700,width=8000,height=5000,font_path='simhei.ttf',random_state=10)
wc.generate(poststr)
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
#读取数据库中的评论
sqlcomment = "SELECT content FROM 21_dogesoft_comment"
cursor.execute(sqlcomment)
resultcomment = cursor.fetchall()
```

```python
dfcomment = pd.DataFrame(list(resultcomment), columns = ['content'])
```

```python
#将评论中的姓名和“回复”“Reply to”进行替换
i = 0
while i < dfemployee.shape[0]:
    if dfemployee.iloc[i,3] is None:
        i = i + 1
    else:
        dfcomment['content'] = dfcomment['content'].apply(lambda x: x.replace(dfemployee.iloc[i,3],' '))
        i = i + 1
i = 0
while i < dfemployee.shape[0]:
    if dfemployee.iloc[i,1] is None:
        i = i + 1
    else:
        dfcomment['content'] = dfcomment['content'].apply(lambda x: x.replace(dfemployee.iloc[i,1],' '))
        i = i + 1
dfcomment['content'] = dfcomment['content'].apply(lambda x: x.replace('回复',' '))
dfcomment['content'] = dfcomment['content'].apply(lambda x: x.replace('Reply to',' '))                                                  
```
